```python
# %load ds_imp_set.py
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.set_printoptions(suppress=True, precision=4)
pd.set_option('display.max_columns', None)
sns.set_context(context='notebook')
sns.set_style(style='whitegrid')

seed = 23
```

```python
from imblearn.over_sampling import RandomOverSampler
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.model_selection import KFold, cross_val_score, cross_val_predict, train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

### Load Data

```python
df = pd.read_csv('./csv/diabetic_classification.csv')
X = df.iloc[:, 1:-1]
y = df['developed_diabetes']
```

### Feature Selection

```python
# feature extraction
pca = PCA(random_state=seed)
pca_fit = pca.fit(X)
# summarize components
np.set_printoptions(suppress=True, precision=4)
cum_var_exp = np.cumsum(pca_fit.explained_variance_ratio_)
print('Cumulative explained variance: ', cum_var_exp)
```

```python
pca_fit.explained_variance_ratio_[9:13]
```

The first 11 principal components capture the explained variance for this data set.  
This can be seen in the graph below.

```python
plt.figure(figsize=(9,7))
ax = plt.subplot()
component_number = range(1, len(cum_var_exp) + 1)
plt.plot(component_number, cum_var_exp, lw=1)
plt.axvline(x=11, color='k')
plt.title('Cumulative Explained Variance vs Number of Principal Components')
ax.set_xlabel('Principal Component')
ax.set_ylabel('Cumulative Explained Variance')
plt.show()
```

### Preprocessing
Ressample with random oversampler to balance classes.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,  
                                                    random_state=seed,
                                                    test_size=0.33)
```

```python
ros = RandomOverSampler(ratio='minority', random_state=seed)
logr = LogisticRegression(random_state=seed)
pca = PCA(n_components=11, random_state=seed)
ss = StandardScaler()

X_train_scale = ss.fit_transform(X_train)
X_test_scale = ss.transform(X_test)
```

```python
Z_train = pca.fit_transform(X_train_scale)
Z_test = pca.transform(X_test_scale)
Z_res, y_res = ros.fit_sample(Z_train, y_train)
```

### Modeling

```python
# note it is not appropriate to use Kfold with PCA unless each training fold has separate PCA
model = logr.fit(Z_res, y_res)
```

```python
y_hat = model.predict(Z_test)
```

```python
y_score = model.decision_function(Z_test)
```

### Evaluation

```python
print(classification_report(y_test, y_hat))
```

"true negatives is C_{0,0}, false negatives is C_{1,0}, true positives is C_{1,1} and false positives is C_{0,1}"
