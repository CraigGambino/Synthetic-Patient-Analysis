```python
import pandas as pd
pd.set_option('display.max_columns', None)
df = pd.read_csv('./csv/diabetic_dirty.csv')

df.loc[:, 'Body Height':'Urea Nitrogen'].head()

na_totals = list(df.loc[:, 'Body Height':'Urea Nitrogen'].isnull().sum().unique())

na_totals
```

```python
df.head()
```

```python
df.loc[:, 'Body Height':'Urea Nitrogen'].isnull().sum()
```

```python
na_counts = dict(df.loc[:, 'Body Height':'Urea Nitrogen'].isnull().sum())
```

```python
na_groups = [tuple(col for col,number in na_counts.items() if number==total) for total in na_totals]
na_groups
```

```python
[(len(group), dict(df.loc[:,group].isnull().sum(axis=1).value_counts())) for group in na_groups]
```

```python
(df['Calcium'].notnull() & df['Body Height'].isnull()).any()
```

```python
all_null = list(df[df['Calcium'] \
               .isnull()] \
                .index)
```

```python
df.drop(all_null, axis=0, inplace=True)
df.index = list(range(df.shape[0]))
```
